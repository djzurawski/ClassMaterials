# Use Pandas to plot the $CO_2$ data from the previous exercise

```python
%matplotlib inline   

import urllib2
import StringIO 
import itertools 
import datetime
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
url = "ftp://aftp.cmdl.noaa.gov/products/trends/co2/co2_mm_gl.txt"
```

```python
names = ['month','day','year_dec','average','trend']
df = pd.read_csv(url, comment='#',header=None, sep='    ',names=names)
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['year_dec'], df['average'],'.-', color='r')
plt.plot(df['year_dec'], df['trend'],'.-', color='k')
plt.title("Recent Global Monthly Mean CO2", fontsize=14)
plt.xlabel("Year")
plt.ticklabel_format(useOffset=False)
plt.xlim([2011,2016])
plt.xticks([x for x in xrange(2011,2017)])
plt.ylim([385,405])
plt.ylabel("Parts Per Million")
plt.show()
```
